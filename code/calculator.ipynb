```python
import pandas as pd
import numpy as np
data = pd.read_csv("E:\KPI\Databases\Lab2\source csv\penguins_lter.csv")
# sizes = pd.read_csv("penguins_size.csv")
data = data.rename(columns={'studyName':'Study Name'})
data = data.drop(columns=['Delta 15 N (o/oo)', 'Delta 13 C (o/oo)', 'Comments', 'Individual ID'])
# sizes

data['Study Author'] = 'Dr. Kristen Gorman'
data['Study Program'] = 'Palmer Station Long Term Ecological Research Program'

data_islands = pd.DataFrame(data['Island'])
data_islands.set_index('Island', inplace=True)
data_islands.loc['Torgersen', 'Longtitude'] = -64.083333
data_islands.loc['Torgersen', 'Latitude'] = -64.766667
data_islands.loc['Dream', 'Longtitude'] = -64.233333
data_islands.loc['Dream', 'Latitude'] = -64.733333
data_islands.loc['Biscoe', 'Longtitude'] = -65.5
data_islands.loc['Biscoe', 'Latitude'] = -65.433333
data_islands = data_islands.drop_duplicates()

data = pd.merge(data, data_islands, left_on='Island', right_on='Island')

def to_snake_case(line):
    line = line.lower()
    line = line.replace(" ", '_')
    return line

def to_bool(line):
    if line == 'Yes':
        return "TRUE"
    elif line == 'No':
        return "FALSE"
    else:
        return 'NULL'

def wrap_string(line):
    return "'" ++ line.to_string() ++ "'"

def to_string(line, skip_wrapping):
    string = []
    for index, value in line.iteritems():
        if (index in skip_wrapping) or (value == "NULL"):
            string.append(str(value))
        else:
            # print(value)
            string.append("'" + str(value) + "'")
    return ", ".join(string)

def to_datetime(line):
    string = line.split('/')
    return '20'+string[2]+'-'+string[0]+"-"+string[1]

data.rename(columns={"Culmen Length (mm)":'Culmen Length', 'Culmen Depth (mm)':'Culmen Depth', 'Flipper Length (mm)':'Flipper Length', 'Body Mass (g)':'Body Mass' }, inplace=True)
data.columns = pd.Series(data.columns).apply(to_snake_case)
data['clutch_completion'] = data['clutch_completion'].apply(to_bool)
data.rename(columns={'study_name':'study_number', 'island':'island_name'}, inplace=True)
data.fillna("NULL", inplace=True)
data['date_egg'] = data['date_egg'].map(to_datetime)


display(data)
# display(data['Species'].unique())
# display(data['Island'].unique())
# display(data[(data["Island"] == 'Torgersen')])
display(data.columns)
data.to_csv("E:\KPI\Databases\Lab2\source csv\penguins_clear.csv")
```

```python

numbers_penguin = ['sample_number', 'clutch_completion', 'culmen_length', 'culmen_depth', 'flipper_length', 'body_mass']
penguin_record = data.drop(columns=['region', 'longtitude', 'latitude', 'study_author', 'study_program'])
indexes_pengiun_record = ", ".join(list(penguin_record.columns))

islands = data[['island_name', 'longtitude', 'latitude', 'region']].drop_duplicates()
islands.reset_index(inplace=True)
islands.drop(columns='index', inplace=True)
numbers_island = ['longtitude', 'latitude']
indexes_islands = ", ".join(list(islands.columns))

studies = data[['study_number', 'study_author', 'study_program']].drop_duplicates()
studies.reset_index(inplace=True)
studies.drop(columns='index', inplace=True)
numbers_study = []
indexes_studies = ", ".join(list(studies.columns))

with open('E:\KPI\Databases\Lab2\sql_populate_generated.sql', 'w') as file:
    for i in range(len(islands)):
        file.write('INSERT INTO Islands(')
        file.write(indexes_islands)
        file.write(')\nValues(')
        file.write(to_string(islands.loc[i], numbers_island))
        file.write(');\n')

    file.write('\n\n------------------\n\n')

    for i in range(len(studies)):
        file.write('INSERT INTO Studies(')
        file.write(indexes_studies)
        file.write(')\nValues(')
        file.write(to_string(studies.loc[i], numbers_study))
        file.write(');\n')

    file.write('\n\n------------------\n\n')

    for i in range(len(penguin_record)):
        file.write('INSERT INTO PenguinRecords(')
        file.write(indexes_pengiun_record)
        file.write(')\nValues(')
        file.write(to_string(penguin_record.loc[i], numbers_penguin))
        file.write(');\n')



```
